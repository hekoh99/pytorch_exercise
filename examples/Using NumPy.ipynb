numpy는 computation graph나 딥러닝, gradient에 대해서 직접적인 동작을 정의하는 함수나 메서드는 없지만 numpy를 사용해 신경망의 순전파 단계와 역전파 단계를 직접 구현할 수 있음

```python
# 3차 다항식이 sine 함수에 근사하도록 만들기
import numpy as np
import math
```

```python
# 무작위로 입력과 출력 데이터 생성

x = np.linspace(-math.pi, math.pi, 2000)
y = np.sin(x)
print(x.shape)
print(y.shape)
```

```python
# 무작위로 가중치 초기화
a = np.random.randn()
b = np.random.randn()
c = np.random.randn()
d = np.random.randn()
```

```python
learning_rate = 1e-6
point_num = x.shape[0]
for t in range(point_num):
  # 순전파 단계 : 에측값을 계산함
  y_pred = a + b * x + c * x ** 2 + d * x **3
  loss = np.square(y_pred - y).sum()
  if t % 100 == 99 :
    print(t, loss)
```
