```python
import numpy as np
import open3d as o3d
import os
import sys

dir = "/content/structure-from-motion/SfM_quality_evaluation/Benchmarking_Camera_Calibration_2008/entry-P10/points/"
file_format = "_images.ply"
files = next(os.walk(dir))[2]
print(files)
```

```python
from pyntcloud import PyntCloud

for file in files :
  file = dir + file
  point_cloud = PyntCloud.from_file(file)
  point_cloud.plot()
```

```python
def main():
    cloud = o3d.io.read_point_cloud(file) # Read the point cloud
    o3d.visualization.draw_geometries([cloud]) # Visualize the point cloud     

if __name__ == "__main__":
    main()
```
