```python
import mmcv
from terminaltables import AsciiTable
from mmdet.apis import init_detector, inference_detector, show_result_pyplot, show_result_ins
import pickle
import os
import sys

config_file = './configs/solo/decoupled_solo_r50_fpn_8gpu_3x.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '/content/drive/MyDrive/Colab_Notebooks/pytorch/code/lab/plenoptic/pretrained/DECOUPLED_SOLO_R50_3x.pth'

# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')

category_list = ['/taekgyun2', '/taekgyun1', '/dance', '/maskdance']

IMAGE_DIR = '/content/drive/MyDrive/Colab_Notebooks/pytorch/code/lab/plenoptic/data/video_capture'
category = '/taekgyun2'
IMAGE_DIR = IMAGE_DIR + category

file_names = next(os.walk(IMAGE_DIR))[2]
file_names = file_names[2:]
print(file_names)
```

```python
result_dir = '/content/drive/MyDrive/Colab_Notebooks/pytorch/code/lab/plenoptic/data/model_output/SOLO_output'
visual_dir = '/content/drive/MyDrive/Colab_Notebooks/pytorch/code/lab/plenoptic/data/visualization/solo_result'
for file in file_names:
  img = IMAGE_DIR + '/' + file
  result_file = result_dir + category + '/result_' + file[:-4] + '.pickle'
  result = inference_detector(model, img)
  visual_file = visual_dir + category + '/result_' + file
  # with open(result_file, "wb") as fw:
  #   pickle.dump(result, fw)
  show_result_ins(img, result, model.CLASSES, score_thr=0.25, out_file=visual_file)
  
```
