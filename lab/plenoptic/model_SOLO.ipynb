```python
import mmcv
from terminaltables import AsciiTable
from mmdet.apis import init_detector, inference_detector, show_result_pyplot, show_result_ins

config_file = '../configs/solo/decoupled_solo_r50_fpn_8gpu_3x.py'
# download the checkpoint from model zoo and put it in `checkpoints/`
checkpoint_file = '/content/drive/MyDrive/Colab Notebooks/pytorch/code/lab/plenoptic/pretrained/DECOUPLED_SOLO_R50_3x.pth'

# build the model from a config file and a checkpoint file
model = init_detector(config_file, checkpoint_file, device='cuda:0')

# test a single image
img = '/content/drive/MyDrive/Colab Notebooks/pytorch/code/lab/plenoptic/data/test_data/014.png'
result = inference_detector(model, img)

show_result_ins(img, result, model.CLASSES, score_thr=0.25, out_file="/content/drive/MyDrive/Colab Notebooks/pytorch/code/lab/plenoptic/data/014_solo_result.png")
```
